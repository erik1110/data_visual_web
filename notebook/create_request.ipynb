```python
import requests
from faker import Faker
import json
fake = Faker()
fake = Faker('zh_TW')
```

```python
fake_data = fake.simple_profile()
my_params = {"user_id": fake.ssn(),
             "username": fake_data['username'],
             "name": fake_data['name'],
             "sex": fake_data['sex'],
             "address": fake_data['address'],
             "mail": fake_data['mail'],
             "birthday": fake_data['birthdate'].strftime('%Y-%m-%d'),
             "country": fake.country(),
             "job": fake.job(),
             "phone_number": fake.phone_number(),
             "date": fake.date()}
my_params
```

```python
json.dumps(my_params, ensure_ascii=False)
```

```python
my_params
```

```python
# 將查詢參數加入 GET 請求中
r = requests.post('http://127.0.0.1:5000/create_data', params = my_params)
r
```

```python
# 壓力打 api
for i in range(10000):
    print("="*50)
    print(f"第{i}個")
    fake_data = fake.simple_profile()
    my_params = {"user_id": fake.ssn(),
                 "username": fake_data['username'],
                 "name": fake_data['name'],
                 "sex": fake_data['sex'],
                 "address": fake_data['address'],
                 "mail": fake_data['mail'],
                 "birthday": fake_data['birthdate'].strftime('%Y-%m-%d'),
                 "country": fake.country(),
                 "job": fake.job(),
                 "phone_number": fake.phone_number(),
                 "date": fake.date()}
    # 將查詢參數加入 GET 請求中
    r = requests.post('http://127.0.0.1:5000/create_data', params = my_params)
    print("status:", r)
```
